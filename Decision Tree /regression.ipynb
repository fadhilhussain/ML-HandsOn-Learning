```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
print(dataset['target'])
```

## DataFrame 

```python
import pandas as pd 
df = pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df
```

## independent and dependent feature 

```python
X = df
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train
```

```python
y_train
```

###  check correlation 

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,7))
sns.heatmap(X_train.corr(),annot=True)
```

### model training 

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor() # cheythikkilla cheyynam , direct hyper ilek 
regressor.fit(X_train,y_train)
```

### Hyperparameter tuning 

```python
params = {
    'criterion' : ["squared_error","friedman_mse","absolute_error"],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5,10,15,20,25],
    'max_features' : ['auto','sqrt','log2']
}
treemodel = DecisionTreeRegressor()
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(treemodel, param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
