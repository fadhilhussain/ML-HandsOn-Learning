### 4,  data cleaning 


```python
df.isnull().sum()

#dfroping all missing valus rows
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#duplicated data

df[df.duplicated()]
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
#dealing with data type 

df.info()
```

```python
df['rating'].unique()
```

```python
#rating 

df['rating']= df['rating'].replace({'No rating':0, 'New ':np.nan})
df['rating'].unique()
df['rating'] = df['rating'].astype(float)
df['rating'] = df['rating'].fillna(df['rating'].mean())
df[df['rating'].isnull()]
```

```python
df.info()
```

```python
df.head()
```

```python
#bedrooms 
df['baths'].unique()
df['baths'] = df['baths'].replace('Not specified',np.nan)
df['baths'] = df['baths'].astype(float)
df['baths'] = df['baths'].fillna(1.)
```

```python
df['baths'].unique()
```

```python
numerical_feature = [feature for feature in data.columns if data[feature].dtype!= 'O'][4:]
plt.figure(figsize=(15,6))
plt.suptitle('Univariate Analysis of numerical feature')
for i in range(0,len(numerical_feature)):
    plt.subplot(2,5,i+1)
    sns.histplot(data = data, x=df[numerical_feature[i]])
    plt.xlabel(numerical_feature[i])
    plt.tight_layout()
```
