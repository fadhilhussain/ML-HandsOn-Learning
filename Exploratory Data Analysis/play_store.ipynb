```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
#1
#convert into object type reviews into integer type 
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
#2 size 
df_copy['Size'].unique()
```

19000K = 19M

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size'].unique()
```

```python
df_copy.info()
```

```python
df_copy['Size'] # nan maattnam feature engineer nn chyyya
```

```python
df_copy.info()
```

```python
#3
#install
df_copy['Installs'].unique()
```

```python
#4
#price
df_copy['Price'].unique()
```

```python
#change special character to blank 
chars_to_remove = ['+','$',',']
column_df = ['Installs','Price']
for item in chars_to_remove:
    for cols in column_df:
        df_copy[cols] = df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.head()
```

```python
#5
#last updated feature 
df_copy['Last Updated'].unique()
```

```python
#convert to it date time 
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['day'] = df_copy['Last Updated'].dt.day
df_copy['month'] = df_copy['Last Updated'].dt.month
df_copy['year'] = df_copy['Last Updated'].dt.year
```

## internal assignment 

  1.  Which category has largest number of installation?
  2.  What are the top 5 most intalled Apps in each popular categories ?
  3.  How many apps are the there on google play store which get 5 ratings?

```python
df_copy.head()
```

```python
#1 
categori_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
```

```python
categori_installs.Installs =   categori_installs.Installs/1000000000 # convert into billion 
df2 = categori_installs.head(10)
plt.figure(figsize=(14,10))

sns.set_context('talk')
sns.set_style('darkgrid')
ab = sns.barplot(x = 'Installs', y = 'Category', data = df2)
ab.set_xlabel('No of installs in billions ')
ab.set_ylabel('')
ab.set_title('Most popular installed category in playstore' , size = 20)

plt.show()
```

```python
#2
top_5 = df_copy.groupby(['Category','App'])['Installs'].sum().sort_values(ascending = False).reset_index()
category11 = ['GAME','COMMUNICATION','TOOLS','PRODUCTIVITY']
sns.set_context('poster')
sns.set_style('darkgrid')

plt.figure(figsize=(40,30))

for i,app in enumerate(category11):
    df2 = top_5[top_5.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(x='Installs', y='App' , data= df3)
    plt.xlabel('no of installs in billions')
    plt.ylabel('')
    plt.title(app,size = 20)

plt.tight_layout()

plt.show()
```

## Insight

 1. Sunway Surfers are most intalled apps in Game
 2. Skype, whatsapp are popular in commuination
 3. Google lead on TOOLS
 4. Google drive are play on productivity 

```python
rating = df_copy.groupby(['Category','Installs','App'])['Rating'].sum().sort_values(ascending=False).reset_index()
topFiv = rating[rating.Rating==5.0]
print(f'there are {topFiv.shape[0]}  apps are gets 5 ratings ')
topFiv.head(1)
```
