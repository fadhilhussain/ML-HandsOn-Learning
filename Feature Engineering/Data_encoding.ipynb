# DATA ENCODING

Data encoding is the process of converting categorical data (like text or labels) into numerical form so that machine learning algorithms can use it to make predictions

     CATERGORICAL  ----------------------------------------------------> NUMERICAL 
         DATA                       CONVERTED IN TO                         DATA

         THIS PROCESSES IS KNOWN AS DATA ENCODING

# LABEL ENCODING

                                ASSIGN A UNIQUE NUMERICAL LABEL TO EACH 
                                CATEGORY IN THE VARIABLE 

                                     RED -------> 1
                                    GREEN ------> 2
                                     BLUE ------> 3

DISADVANTAGE OF LABEL ENCODING 

LABEL ENCODING ASSIGNS A NUMBER TO EACH CATEGORY. THE PROBLEM IS THAT IT CAN MAKE THE MODEL THINK THERE'S AN ORDER OR RANKING BETWEEN THE CATEGORIES (LIKE 0 < 1 < 2), EVEN WHEN THERE ISN'T. THIS CAN CONFUSE THE MODEL AND AFFECT ITS PREDICTIONS, ESPECIALLY WITH NON-ORDERED CATEGORIES LIKE COLORS OR NAMES.



```python
data = pd.DataFrame({
    'color' : ['red','red','blue','green','blue','green']
})
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encoder = LabelEncoder()
labels = lbl_encoder.fit_transform(data[['color']])
df_labels = pd.DataFrame(labels)
df_labels
```

```python
#adding a value 
lbl_encoder.transform([['red']])
```

```python
df_labels.value_counts()
```

# ORDINAL ENCORDING 

          USED TO ENCODE CATEGORICAL DATA THAT HAVE AN INTRISNIC ORDER OR RANKING , 
          EACH CATEGORY IS ASSIGNED AA NUMERICAL VALUE BASED ON ITS POSITION IN THE ORDER 

          KIDS - 0
          TEENAGE - 1                    HERE ABOVE 50  HAVE MORE PRIORITY THAN KIDS
          ADULTS - 2
          ABOVE 50 - 3 

          ASSIGNING VALUE BASED ON ITS RANKING 

```python
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({
    'size' : ['small','medium','large','medium','small','large']
})
df
```

```python
#assing ranking 
encoder = OrdinalEncoder( categories = [['small','medium','large']])

df_e = encoder.fit_transform(df[['size']])
df_e
```

```python
df_label = pd.DataFrame(df_e)

pd.concat([df,df_label] , axis= 1)
```

```python
encoder.transform([['small']])
```

# TARGET GUIDED ORDINAL ENCORDING 

 ITS A TECHINIQUE USED TO ENCODE CATEGORICAL VARIABLE BASED ON THEIR RELATIOSHIP WITH TARGET VARIABLE 

 USEFULL WHEN WE HAVE LARGE NUMBER OF UNIQUE CATEGORY 

 ITS WORK BASED ON THE MEAN OR MEDIAN OF THE TAGRET VARIABLE FOR THAT CATEGORY

 

```python
df1 = pd.DataFrame({
    'city' : ['yemen','ksa','dubai','yemen','dubai','ksa'],
    'price' : [200,340,290,120,500,340]
})
df1
```

#### here price is the target variable , we convert city categorical variable to numerical variable based
#### on imputatons of target variable 

```python
# find the mean of prize based on city 
mean_price = df1.groupby('city')['price'].mean()
mean_price
```

```python
# change to dictionary
mean_price = df1.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df1['city_encoded'] = df1['city'].map(mean_price)
df1['city_encoded']
```

```python
#for training ,
df1[['city_encoded','price']]    # remove city column 
```

## ADVANTAGE AND DISADVANTAGE 

### Advantages of Target Ordinal Encoding:
#### * Captures Ordinal Relationships:
It preserves the ordinal relationships between categories, making it suitable for features where the order matters (e.g., rankings).
#### * Model Interpretability:
Since the encoded values represent the target's mean, it can make the model more interpretable as the encoding is directly related to the target variable.
#### * Effective for Tree-based Models:
Tree-based models like decision trees and random forests can handle target ordinal encoding well, as they can split based on the encoded values, capturing meaningful relationships.
#### * Reduction of Dimensionality: 
It reduces the dimensionality of categorical features, as it converts categories into numerical values rather than creating many columns like one-hot encoding.
### Disadvantages of Target Ordinal Encoding:
#### * Risk of Overfitting:
Since the encoding is directly related to the target variable, it can lead to overfitting, especially with small datasets or high cardinality features.
#### * Data Leakage:
If the encoding is done on the entire dataset (including test data), it can lead to data leakage. The target information might accidentally influence the encoding, making the model's evaluation metrics unreliable.
#### * Not Suitable for All Models:
Some models, especially linear models, may not benefit from target ordinal encoding, as it can introduce linear relationships that do not exist in reality.
#### * Complexity:
Implementing target ordinal encoding requires careful handling, such as using cross-validation to mitigate overfitting and data leakage.
