## IMPUTATION BY FORWARD AND BACKWARD FILLING 

```python
import numpy as np
data = {
    'name' : ['fadhil','qinan','jaseem','raheem','fayad','ismail'],
    'age' : [21,22,np.nan,21,np.nan,22]
}
df = pd.DataFrame(data)
df
```

```python
df['age']
```

```python
df['ffill']=df['age'].fillna(df['age'].ffill())
```

```python
df['bfill']=df['age'].fillna(df['age'].bfill())
```

```python
df[['ffill','bfill']]
```

```python
df['ffill'].value_counts()
```
