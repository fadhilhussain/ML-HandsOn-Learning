```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data['data'],columns=data['feature_names'])
df.head()
```

## standardization 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

## Applying PCA algorithm 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
dat_pca = pca.fit_transform(scaled_data)
```

```python
dat_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(dat_pca[:,0],dat_pca[:,1],c = data['target'],cmap='plasma')
plt.xlabel('First principle componets')
plt.ylabel('Second principle componets')
plt.show()
```
