```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn import datasets
```

```python
data = datasets.load_digits()
```

```python
data
```

```python
df = pd.DataFrame(data=data['data'],columns=data['feature_names'])
df['target'] = data['target']
```

```python
df
```

```python
X = df.drop(columns = 'target',axis=1)
y = df['target']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
data_scaled = scaler.transform(X)
```

```python
data_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
dat_pca = pca.fit_transform(data_scaled)
```

```python
dat_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(dat_pca[:,0],dat_pca[:,1],c = data['target'],cmap='plasma')
plt.colorbar(scatter, label="target") 
plt.xlabel('First principle componets')
plt.ylabel('Second principle componets')
plt.show()
```

```python
df1 = pd.DataFrame(dat_pca,columns=['pca1','pca2'])
df1['y'] = df['target']
df1['pca1*pca2'] = (df1['pca1'] * df1['pca1'])
df1['pca1_square'] = df1['pca1']**2
df1['pca2_square'] = df1['pca2']**2
```

```python
df1
```

```python
import plotly.express as px
plt.figure(figsize=(15,8))
fig = px.scatter_3d(df1,x='pca1', y='pca2', z='pca1*pca2',color = 'y')
fig.show()
```

```python
plt.figure(figsize=(15,7))
fig = px.scatter_3d(df1,x='pca1_square', y='pca2_square', z='pca1*pca2',color = 'y')
fig.show()
```
