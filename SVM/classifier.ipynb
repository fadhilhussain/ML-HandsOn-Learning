```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#create datapoints 
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

## Linear SVC 


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC 
```

#### when data points are linearly seprable we can use kernal as ''' LINEAR '''
#### '' rbf '' etc...  keranl mathemtic formula use cheyth hiegher degree ilek maaty iytt accuracy koottam new plane kittum for clearly seprable data sets 

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python

#preddiction 
y_pred = svc.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train,y_train)
```

```python
y_pred1 = rbf.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train,y_train)
```

```python
y_pred2 = poly.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
```

```python
y_pred3 = sigmoid.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

## Hyperparameter tuning 

```python
from sklearn.model_selection import GridSearchCV

#params 
param = {
    'C' : [0.1,1,10,100,1000],
    'gamma' : [1,0.1,0.01,0.001,0.0001],
    'kernel' : ['rbf']
}
```

```python
grid = GridSearchCV(SVC(),param_grid = param,refit=True,cv=5,verbose=3)
```

### refit :- in every cross validation it will refit the model  [oro iteration ilum refit cheyyum ]

```python
grid.fit(X_train,y_train)
```
