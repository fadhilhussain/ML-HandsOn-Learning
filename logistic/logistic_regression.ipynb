### Hyperparameter Tuning and cross validation

## Grid Search Cv 

```python
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

## Randomized Search Cv

```python
model = LogisticRegression()
from sklearn.model_selection import RandomizedSearchCV
randomcv = RandomizedSearchCV(estimator = model, param_distributions = params , scoring = 'accuracy', cv = 5)
```

```python
randomcv
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_predr  = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_test,y_predr)
print(score)
cm = confusion_matrix(y_test,y_predr)
print(cm)
print(classification_report(y_test,y_predr))
```

### hyperparameter tuning 

```python
model1 = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c = [0.001, 0.01, 0.1, 1, 10]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
iter = [100,200,300,350]
multiclass = ['ovr','multinomial']
parameter = dict(penalty=penalty,C=c,solver=solver,max_iter=iter,multi_class=multiclass)
```

```python
parameter
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
from sklearn.model_selection import GridSearchCV
grid2 = GridSearchCV(estimator=model1,param_grid=parameter,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid2
```

```python
grid2.fit(X_train,y_train)
```

```python
grid2.best_score_
```

```python
grid2.best_params_
```

```python
y_predict = grid2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_predict)
cm = confusion_matrix(y_test,y_predict)
print(score)
print(cm)
print(classification_report(y_test,y_predict))
```
