```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y=True)
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## TRAIN TEST SPLIT 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train
```

```python
labelgender = LabelEncoder()
labelsmoker = LabelEncoder()
```

```python
X_train['sex'] = labelgender.fit_transform(X_train['sex'])
X_train['smoker'] = labelsmoker.fit_transform(X_train['smoker'])
```

```python
X_test['sex'] = labelgender.transform(X_test['sex'])
X_test['smoker'] = labelsmoker.transform(X_test['smoker'])
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('onehot',OneHotEncoder(drop='first'),[4])],remainder='passthrough')
```

```python
import sys
import numpy 
numpy.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
X_train
```

## naive bayes 


```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
