```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('multiple_linear_regression_dataset.csv')
df
```

```python
df.corr()
```

```python
import seaborn as sns 
```

```python
sns.pairplot(df)
```

```python
df.isnull().sum()
```

```python
## visualize more closely 
plt.scatter(df['experience'],df['income'],color = 'red')
plt.xlabel('experience')
plt.ylabel('income')
```

```python
#independent variable and dependent variable 
X = df.iloc[:,:-1]                              # X = df[['age','experience']]
y = df.iloc[:,-1]                               # y = df['income']
```

```python
# train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 42)
```

## regression plot 

```python

sns.regplot( x = 'age', y = 'experience' , data = df)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
#linear regrssion
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

### cross validation

```python

from sklearn.model_selection import cross_val_score
validation_score = cross_val_score(regression,X_train,y_train,scoring= 'neg_mean_squared_error',cv = 3)
```

```python
validation_score
```

```python
#finding avearge 
np.mean(validation_score)
```

```python
## predication

y_pred = regression.predict(X_test)
y_pred
```

```python
#performance metrix 

from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
adjusted_r2 = 1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print(f" r2 = {score} and adjusted r2 = {adjusted_r2}")
```
