```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('ice_cream.csv')
```

```python
df
```

```python
#independent and dependent feature 

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#splitting  datas 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.06,random_state=42)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    poly_feature = PolynomialFeatures(degree = degree, include_bias= True)
    lnr_reg = LinearRegression()
    poly_regression = Pipeline([('poly_feature',poly_feature),('lnr_reg',lnr_reg)])
    poly_regression.fit(X_train,y_train)
    y_new = poly_regression.predict(X_test)
    #plotting 
    plt.scatter(X_test,y_new,color='r')
    plt.scatter(X_train,y_train,color='g')
```
