```python
df.drop(122,inplace = True)
```

### make a new columns region and assign

  2. 1 :- sidi-bell abbes

```python
df = df.dropna().reset_index(drop=True)
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
#fix spaces in column name 
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## change required column as integer datatype 


```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

## changing other columns to float data type


```python
object = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
object
```

```python
for i in object:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
#save the clean dataset
df.to_csv('Algerian_cleaned.csv',index=False)
```

## Exploratory data analysis

```python
#drop day, month, and year
data = df.drop(['day','month','year'],axis=1)
data
```

```python
#encoding all categoreis in classes 
data['Classes'] = np.where(data['Classes'].str.contains('not fire'),0,1)
```

```python
data
```

```python
#categoreis in classes 
data['Classes'].value_counts()
```

```python
data.head()
```

```python
data['Classes'].value_counts()
```

### density plot of all features 

```python
data.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
print(plt.style.available)
```

### percentage of pie chart 

```python
percentage = data['Classes'].value_counts(normalize=True)*100  # normalize nokiyal percentage value kittum 
percentage
```

```python
#plotting pie chart 

classslabel = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels = classslabel,autopct='%1.f%%')
plt.title('Pie chart of Classes')
plt.show()
```
